### About Dataset
#### Context
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage

#### Content
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

#### Acknowledgements

### DATA CLEANING
I noticed some zeros so i need to research on that and clean accordingly. It is not possible for Glucose, BloodPressure, SkinThickness, Insulin and	BMI to have '0' as its value. I would convert the '0' values to NAN and explore using the python missingno library.

```python
missing_values= diabetes.loc[:,'Glucose':'BMI']
missing_values.replace(0, np.nan, inplace=True)
diabetes.loc[:,'Glucose':'BMI'] = missing_values
```

##### Missingno Heatmap
The nullity correlation ranges from -1 to 1.

1. -1 - Exact Negative correlation represents that if the value of one variable is present then the value of other variables is definitely absent.
2. 0 - No correlation represents that variables values present or absent do not have any effect on one another.
3. 1 - Exact Positive correlation represents that if the value of one variable is present then the value of the other is definitely present.

We can see from the heatmap nelow that Insulin and SkinThickness have a correlation of 0.7 which is close to 1, and so if an Insulin value is missing then there is a very high chance for SkinThickness value to be missing as well.

```python
msno.heatmap(diabetes)
```

```python
# drop rows having missing values for Glucose or BMI since they are few
diabetes.dropna(subset=['Glucose','BMI'],inplace=True)
```

Using KNN Imputer to fill thr missing values for BloodPressure, SkinThiickness and Insulin

```python
from sklearn.impute import KNNImputer
```

```python
db_clean=diabetes.copy()

imputer=KNNImputer(n_neighbors=10)

db_clean.iloc[:]=imputer.fit_transform(db_clean)
```

```python
# Checking to see if there are any null values left
db_clean.isnull().sum()
```

```python
db_clean.describe()
```

#### Let's take a look out the outliers using boxplots

```python
#Outliers in pregnancies column
sns.boxplot(data=db_clean,x='Pregnancies')
```

Pregnancy values from about 13 and above are outliers, so we remove them:

```python
db_clean=db_clean[db_clean['Pregnancies']<=13]
```

```python
#Outliers in glucose column
sns.boxplot(data=db_clean,x='Glucose')
```

Glucose has no outliers.

```python
#Outliers in BloodPressure column
sns.boxplot(data=db_clean,x='BloodPressure')
```

Glucose values below about 35 and above about 110 are outliers, so we remove them:

```python
db_clean=db_clean[(db_clean['BloodPressure']>=35) & (db_clean['BloodPressure']<=110)]
```

```python
#Outliers in SkinThickness column
sns.boxplot(data=db_clean,x='SkinThickness')
```

SkinThickness values above about 65 are outliers, so we remove them:

```python
db_clean=db_clean[db_clean['SkinThickness']<=65]
```

```python
#Outliers in Insulin column
sns.boxplot(data=db_clean,x='Insulin')
```

Insulin values above about 500 are outliers, so we remove them:

```python
db_clean=db_clean[db_clean['Insulin']<=500]
```

```python
#Outliers in Age column
sns.boxplot(data=db_clean,x='Age')
```

Age values above about 75 are outliers, so we remove them:

```python
db_clean=db_clean[db_clean['Age']<=75]
```

```python
db_clean.describe()
```
